```python
# Function to identify  if tokens are named entities and replace them with UNKNOWN
def remove_details(word):
  if word.pos_ =='VERB':
    return ' <mask> '
  return word.string


# Function where each token of spacy doc is passed through remove_deatils()
def update_article(doc):
  # iterrating through all entities
  for ent in doc.ents:
    ent.merge()
  # Passing each token through remove_details() function.
  tokens = map(remove_details,doc)
  return ''.join(tokens)

# Passing our news_doc to the function update_article()
my_doc1 = update_article(my_doc)
```

```python
from textblob import TextBlob
df1['label'] = df1['sentence'].apply(lambda Tweet: TextBlob(Tweet).sentiment)
print(df1)
```

```python
df2=pd.DataFrame(df1['label'].tolist(), index= df1.index)
```

```python
df2.head()
```

```python
df_new = df1
df_new['polarity'] = df2['polarity']
df_new.polarity = df2.polarity.astype(float)
df_new['subjectivity'] = df2['subjectivity']
df_new.subjectivity = df2.polarity.astype(float)
print(df_new)
```

```python
def met_multiclass(score):
  score = float(score)
  if score == 0:
    return 0
  elif score < 0:
    return 0
  elif score > 0: 
    return 1

df_new['label'] = df_new.polarity.apply(met_multiclass)
```
