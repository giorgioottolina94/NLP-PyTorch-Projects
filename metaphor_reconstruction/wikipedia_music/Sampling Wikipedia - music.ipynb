```python
from tqdm.notebook import tqdm
import wikipedia
# see https://wikipedia.readthedocs.io/en/latest/code.html
wikipedia.set_lang("en")
seeds = ("music", "science", "poetry")
summaries = []
for seed in seeds:  
  seed_pages = wikipedia.search(seed, results=500)
  for page in tqdm(seed_pages):
      summary = wiki.page(page).summary
      summaries.append(summary)
```

```python
import pandas as pd
sums_pd = pd.DataFrame({"Original sentences": str(summaries).split('.')})
sums_pd.to_csv("summaries_gr.csv.gz", index=False, compression="gzip")
sums_pd.head()
```

```python
# Function to identify  if tokens are named entities and replace them with UNKNOWN
def remove_details(word):
  if word.pos_ =='VERB':
    return ' <mask> '
  return word.string


# Function where each token of spacy doc is passed through remove_deatils()
def update_article(doc):
  # iterrating through all entities
  for ent in doc.ents:
    ent.merge()
  # Passing each token through remove_details() function.
  tokens = map(remove_details,doc)
  return ''.join(tokens)

# Passing our news_doc to the function update_article()
my_doc1 = update_article(my_doc)
```
