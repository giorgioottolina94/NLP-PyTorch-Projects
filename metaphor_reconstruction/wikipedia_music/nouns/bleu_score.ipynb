Inputs to `score` are a list of candidate sentences and a list of reference sentences. 

```python
with open("stockholm/wikipedia_music/nouns/hyps.txt") as f:
    cands = [line.split() for line in f]

with open("stockholm/wikipedia_music/nouns/refs.txt") as f:
    refs = [line.split() for line in f]
```

Let's have a look.

```python
cands[1]
```

```python
refs[1]
```

```python
import nltk
nltk.translate.bleu_score.corpus_bleu(refs, cands)
```

```python
import nltk
from nltk.translate.bleu_score import SmoothingFunction
smoothie = SmoothingFunction().method4
nltk.translate.bleu_score.corpus_bleu(refs, cands, smoothing_function=smoothie)
```
