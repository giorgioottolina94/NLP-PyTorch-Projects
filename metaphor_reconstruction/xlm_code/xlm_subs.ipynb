#MOH-X XLM

```python
predictions = pd.read_csv("stockholm/xlm_code/mohx_xlm_subs/mohx_sub.csv")
```

```python
predictions = predictions.rename(columns={"label": "prediction"})
```

```python
predictions.head()
```

```python
predictions.shape
```

```python
final = predictions[(predictions['correct_label'] == predictions['prediction'])]
final.shape
```

```python
final.to_csv("mohx_total_correct.csv", index=False)
```

```python
correct_met = predictions[(predictions['correct_label'] == 1) & (predictions['prediction'] ==1)]
correct_met.shape
```

```python
correct_met.to_csv("mohx_tp.csv", index=False)
```

```python
wrong_met = predictions[(predictions['correct_label'] == 1) & (predictions['prediction'] ==0)]
wrong_met.shape
```

```python
wrong_met.to_csv("mohx_fn.csv", index=False)
```

```python
correct_lit = predictions[(predictions['correct_label'] == 0) & (predictions['prediction'] ==0)]
correct_lit.shape
```

```python
correct_lit.to_csv("mohx_tn.csv", index=False)
```

```python
wrong_lit = predictions[(predictions['correct_label'] == 0) & (predictions['prediction'] ==1)]
wrong_lit.shape
```

```python
wrong_lit.to_csv("mohx_fp.csv", index=False)
```
