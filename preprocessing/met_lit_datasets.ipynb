```python
#train met
import pandas as pd
ger_train_met = pd.read_fwf('German/poems/train_met.txt')
ger_train_met = ger_train_met.rename(columns={'jung Tag': 'sentence'})
```

```python
ger_train_met.to_csv('ger_train_met.csv', index=False)
```

```python
#train lit
import pandas as pd
ger_train_lit = pd.read_fwf('German/poems/train_nonmet.txt')
ger_train_lit = ger_train_lit.rename(columns={'ganz Leben': 'sentence'})
```

```python
ger_train_lit['label']=0
```

```python
ger_train_lit
```

```python
ger_train_lit.to_csv('ger_train_lit.csv', index=False)
```

```python
#test met
import pandas as pd
ger_test_met = pd.read_fwf('German/poems/test_met.txt')
ger_test_met = ger_test_met.rename(columns={'schwellend Sommernacht': 'sentence'})
```

```python
ger_test_met['label']=1
```

```python
ger_test_met
```

```python
ger_test_met.to_csv('ger_test_met.csv', index=False)
```

```python
#test lit
import pandas as pd
ger_test_lit = pd.read_fwf('German/poems/test_nonmet.txt')
ger_test_lit = ger_test_lit.rename(columns={'zweit Mal': 'sentence'})
```

```python
ger_test_lit['label']=0
```

```python
ger_test_lit
```

```python
ger_test_lit.to_csv('ger_test_lit.csv', index=False)
```

```python
german_full_test = pd.concat([ger_train_met, ger_train_lit, ger_test_met, ger_test_lit], ignore_index=True)
```

```python
german_full_test.to_csv('german_full_test.csv', index=False)
```
