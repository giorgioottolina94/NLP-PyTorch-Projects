```python
predi = pd.read_csv("mohx_preds.csv")
predi
```

```python
predi['true'] = predi['true'].astype('str')
```

```python
predi['true']=predi['true'].str.replace(',','')
```

```python
predi['true']=predi['true'].str.replace('[','')
```

```python
predi['true']=predi['true'].str.replace(']','')
```

```python
predi['preds'] = predi['preds'].astype('str')
```

```python
predi['preds']=predi['preds'].str.replace('[','')
```

```python
predi['preds']=predi['preds'].str.replace(']','')
```

```python
predi['preds'] = predi['preds'].str.replace(" ","")
predi['true'] = predi['true'].str.replace(" ","")
```

```python
predi["pad_preds"] = predi.preds.astype(str).str.ljust(20, "0")
```

```python
predi["pad_true"] = predi.true.astype(str).str.ljust(20, "0")
```

```python
del predi['preds']
del predi['true']
```

```python
predi = predi.rename(columns={'pad_preds': 'preds', 'pad_true': 'true'})
```

```python
predi
```

```python
predi.shape
```

```python
#Correct Predictions
```

```python
correct_mohx = predi[(predi['preds'] == predi['true'])] #tutte le predizioni corrette (sia metaforiche che letterali)
```

```python
correct_mohx.shape
```

```python
#Wrong Predictions
```

```python
wrong_mohx = predi[(predi['preds'] != predi['true'])] #tutte le predizioni errate (sia metaforiche che letterali)
```

```python
wrong_mohx.shape
```
